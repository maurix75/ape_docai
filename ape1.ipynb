```python
import fire
from pathlib import Path
from typing import Optional
from byaldi import RAGMultiModalModel

def main(
    folder: str = "../pippo/",
    index_name: str = "application",
    model_name: str = "vidore/colpali-v1.2"
) -> None:
    """
    Index documents using a RAG multimodal model.

    Args:
        folder: Input directory path containing documents to index
        index_name: Name for the created index
        model_name: Pretrained model identifier
    """
    input_path = Path(folder).resolve()
    if not input_path.exists():
        raise FileNotFoundError(f"Directory not found: {input_path}")

    rag = RAGMultiModalModel.from_pretrained(model_name, verbose=1)

    rag.index(
        input_path=str(input_path),
        index_name=index_name,
        store_collection_with_index=True,
        overwrite=True,
    )

if __name__ == "__main__":
    fire.Fire(main)
```
